# PROBELM1

#### PART1: DOWNLOADING HOURLY DATA

```python
#FOR INDEX
sp500=yf.Ticker("^GSPC")
sp500.info
```

```python
#should be yyyy-mm-dd format
sp500_his_h=sp500.history(start="2023-01-01",end="2024-02-01",interval="1h")
sp500_his_h
```

```python
nvda=yf.Ticker("NVDA")
nvda_his_h=nvda.history(start="2023-01-01",end="2024-02-01",interval="1h")
nvda_his_h
```

#### PART2: CONVERTING HOURLY TO DAILY

```python
def convert_htod(obj_h):
  obj_h=obj_h.drop("Dividends",axis=1)
  obj_h=obj_h.drop("Stock Splits",axis=1)

  obj_t=obj_h.resample('D').ohlc()
  obj_p=pd.DataFrame()
  obj_p["Open"]=obj_t["Open"].open
  obj_p["High"]=obj_t["High"].high
  obj_p["Low"]=obj_t["Low"].low
  obj_p["Close"]=obj_t["Close"].close
  obj_p["Volume"]=obj_h["Volume"].resample('D').sum()
  obj_p=obj_p.dropna()

  return obj_p
```

```python
nvda_his_d=convert_htod(nvda_his_h)
nvda_his_d
```

```python
sp500_his_d=convert_htod(sp500_his_h)
sp500_his_d
```

#### PART3: DAILY RETURN AND CUMULATIVE RETURNS

##### DAILY

```python
def dailyreturns(obj):
  ans=((obj["Close"]-obj["Open"])/obj["Open"])*100
  return ans
```

```python
nvda_ret=dailyreturns(nvda_his_d)
nvda_ret=nvda_ret.dropna()
nvda_ret
```

```python
sp500_ret=dailyreturns(sp500_his_d)
sp500_ret
```

##### CUMULATIVE


```python
def cumreturns(obj):
  ans=((obj["Close"]-obj["Open"][0])/obj["Open"][0])*100
  # ans=((obj["Close"]-obj["Open"])/obj["Open"])*100
  # ans1=(1+ans).cumprod()-1

  return ans
```

```python
nvda_ret_cumulative=cumreturns(nvda_his_d)
nvda_ret_cumulative
```

```python
sp500_ret_cumulative=cumreturns(sp500_his_d)
sp500_ret_cumulative
```

##### PLOT


```python

plt.subplot(1,2,1)
plt.rcParams["figure.figsize"] = [8.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.title("COMPARISION OF DAILY RETURNS")
plt.xlabel("Date")
plt.ylabel("Daily Returns")
plt.plot(nvda_ret.index,nvda_ret.values,'b',label="NVDIA(STOCK)")
plt.plot(sp500_ret.index,sp500_ret.values,'r',label="S&P500(INDEX)")
plt.legend(loc="upper left")




plt.subplot(1,2,2)
plt.rcParams["figure.figsize"] = [8.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.title("COMPARISION OF CUMULATIVE RETURNS")
plt.xlabel("Date")
plt.ylabel("Cumulative Returns")
plt.plot(nvda_ret_cumulative.index,nvda_ret_cumulative.values,'b',label="NVDIA(STOCK)")
plt.plot(sp500_ret_cumulative.index,sp500_ret_cumulative.values,'r',label="S&P500(INDEX)")
plt.legend(loc="upper left")
```

# PROBLEM 2

#### PART1:DOWNLOAD DATA

```python
Reliance=yf.Ticker("RELIANCE.NS")
Reliance_his=Reliance.history(start="2023-01-01",end="2024-02-01",interval="1d")
Reliance_his.drop("Dividends",axis=1,inplace=True)
Reliance_his.drop("Stock Splits",axis=1,inplace=True)
Reliance_his
```

#### PART2: ADDING A NEW DATE TIME ATTRIBUTE AND PLOTTING

```python
Reliance_his["Day"]=Reliance_his.index.weekday
Reliance_his
```

```python
for i in range(len(Reliance_his["Day"])):
  if(Reliance_his["Day"][i]==0):
    Reliance_his["Day"][i]="Monday"

  if(Reliance_his["Day"][i]==1):
    Reliance_his["Day"][i]="Tuesday"

  if(Reliance_his["Day"][i]==2):
    Reliance_his["Day"][i]="Wednesday"

  if(Reliance_his["Day"][i]==3):
    Reliance_his["Day"][i]="Thrusday"

  if(Reliance_his["Day"][i]==4):
    Reliance_his["Day"][i]="Friday"

  if(Reliance_his["Day"][i]==5):
    Reliance_his["Day"][i]="Saturday"

  if(Reliance_his["Day"][i]==6):
    Reliance_his["Day"][i]="Sunday"

Reliance_his
```

```python
days={"Monday","Tuesday","Wednesday","Thrusday","Friday"}
dict_1={"Monday":0,"Tuesday":0,"Wednesday":0,"Thrusday":0,"Friday":0}

for day in days:
  dict_1[day]=Reliance_his.loc[Reliance_his['Day'] == day]["Close"].mean()


dict_1
```

```python
plt.xlabel("Day of the Week")
plt.ylabel("Closing Price")
plt.scatter(Reliance_his["Day"],Reliance_his["Close"],label="CLOSING PRICES")
bool1=True
for key,value in dict_1.items():
  if(bool1):
    plt.scatter(key,value,c='r',label="AVERAGE CLOSING PRICE")
    bool1=False
  else:
    plt.scatter(key,value,c='r')

plt.legend()
```

**HIGHEST CLOSING PRICE FOR TUESDAY**

#### PART3: CALCULATING THE 7 DAY ROLLING MEAN

```python
#just switching off warning that shows up while slicing
pd.options.mode.chained_assignment = None
```

```python
def rolling_mean(obj,name):
  # obj.drop("Rolling Mean",axis=1)
  obj["Rolling Mean"]=""
  j=0
  for i in range(obj.shape[0]):
    if(i<6):
      #VARIABLE WINDOW SIZE-->commented out because only 7day mentioned
      # print(i,j,obj[name][j:i+1].mean())
      # obj["Rolling Mean"][i]=obj[name][j:i+1].mean()
      obj["Rolling Mean"][i]=np.nan

    elif(i==6):
      # print(i,j,obj[name][j:i+1].mean())
      obj["Rolling Mean"][i]=obj[name][j:i+1].mean()
      j=1
    else:
      # print(i,j,obj[name][j:i].mean())
      obj["Rolling Mean"][i]=obj[name][j:i+1].mean()
      j+=1


```

```python
rolling_mean(Reliance_his,"Close")
#dropping first 6 days data as it is immaterial
Reliance_his.dropna(inplace=True)
Reliance_his
```

```python
plt.title("COMPARISION OF ROLLING MEAN AND CLOSE")
plt.xlabel("DATES")
plt.ylabel("PRICES")
plt.plot(Reliance_his.index,Reliance_his["Close"],label='CLOSE')
plt.plot(Reliance_his.index,Reliance_his["Rolling Mean"],c='r',label='ROLLING MEAN CLOSE')
plt.legend()
```
