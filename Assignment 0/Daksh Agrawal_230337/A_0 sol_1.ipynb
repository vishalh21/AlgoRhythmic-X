# Converting Hourly to Daily

```python
def conv_1h_1d(df):
    conv_df=df.copy()
    conv_df.insert(loc=0, column='Date', value=conv_df.index.date)  #inserting a column of date from datetime index
    conv_df.reset_index(drop=True, inplace=True)                    #dropping original datetime index
    #Grouping on the basis of date, applying appropriate aggregating functions
    conv_df=conv_df.groupby("Date").agg({"Open":'first',"High":'max',"Low":'min',"Close":'last',"Adj Close":'last',"Volume":'sum'})
    return conv_df
```

```python
aapl_conv_1d=conv_1h_1d(apple_df_hr)
```

# Calculating Daily and Cumulative Returns

```python
# Daily returns
def daily_returns(df):
    daily=((df['Adj Close']-df['Adj Close'].shift(1))/df['Adj Close'].shift(1))*100
    daily[0]=0
    return daily

#Cumulative returns
def cum_returns(daily):
    cum=daily.copy()
    for i in range(1,len(cum)):
        cum[i]=((100+cum[i-1])*(1+daily[i]/100))-100
    return cum
```

```python
snp500_daily_returns=daily_returns(snp500_conv_1d)
apple_daily_returns=daily_returns(aapl_conv_1d)
```

```python
snp500_daily_returns
```

```python
apple_daily_returns
```

```python
snp500_cum_returns=cum_returns(snp500_daily_returns)
apple_cum_returns=cum_returns(apple_daily_returns)
```

```python
plt.figure(figsize=(11, 5.5))
plt.plot(snp500_cum_returns,label='S&P500')
plt.plot(apple_cum_returns,label='Apple')
plt.xlabel('Date')
plt.ylabel('Cumulative Return (%)')
plt.title('Cumulative Returns S&P500 vs. Apple')
plt.legend()
```

```python
plt.figure(figsize=(11, 5.5))
plt.plot(snp500_daily_returns,label='S&P500')
plt.plot(apple_daily_returns,label='Apple')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.title('Daily Returns S&P500 vs. Apple')
plt.legend()
```
