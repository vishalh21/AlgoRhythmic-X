FUNCTION FOR CUMULATIVE RETURN

```python
def calcreturn(shares,data,k):
    annual = (data.loc[k,'Close']-data.loc[0,'Close'])
    initial = data.loc[0,'Close']*shares

    total=  initial +annual *shares 
    ret = ((total/initial)-1)*100

    return ret
    
```

AVERAGE CLOSING PRICES


```python
def AveragePrice(t):
    q = 0 
    for i in range(t,len(IRCTC),5):
         AlagDataframe.loc[q] = IRCTC.loc[i]
         q = q+1

    avg_clos_pric= 0
    for i in range(0,len(AlagDataframe)):
        avg_clos_pric = avg_clos_pric+AlagDataframe.loc[i,'Close']

    avg_clos_pric = avg_clos_pric/len(AlagDataframe)
    return avg_clos_pric
```

ROLLING MEAN

```python
def RollingMean(data,kapa):
    mean = 0
   # a =7
    prices = []
    for  i in range(0,len(data)):
        if(i<7):
           prices.append(0)
        else:
            mean = (data.loc[i-7,kapa]+data.loc[i-1,kapa]+data.loc[i-2,kapa]+data.loc[i-3,kapa]+data.loc[i-4,kapa]+data.loc[i-5,kapa]+data.loc[i-6,kapa])/7
            prices.append(mean)
    
    data['Mean']=prices
    return data
```
