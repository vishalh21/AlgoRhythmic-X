```python
import pandas as pd
import numpy as np
import yfinance as yf
import json
import os
```

```python
def volatility_calulator(data):
    # Calculate daily returns
    data['Return'] = data['Adj Close'].pct_change()

    # Calculate daily volatility (standard deviation of daily returns)
    daily_volatility = data['Return'].std()

    # Annualize the daily volatility
    return daily_volatility * np.sqrt(len(data))  #number of trading days in a year
```

```python
# list of nifty100 shares, there are 101 shares
nifty100 = pd.read_csv('ind_nifty100list.csv')
shares = nifty100['Symbol'].tolist()

print("Downloading data")
# i was trying to download all together but didint worked as expected
# download and calculate the volatility of each share form the nifty100 list
volatility = []
for share in shares:
    share_df = yf.download(share+'.NS',
                    start='2023-03-01', # form 1 march 2023 as 1 year form 29 feb 2024 previous rebalancing date
                    # end='2024-02-29',
                    period='1y',
                    interval='1d',
                    progress=True,
                    )
    volatility.append({"Share": share, "Volatility": volatility_calulator(share_df)})

# print the shares with the calulated volatility
volatility1 = volatility.copy()
volatility_indented = json.dumps(volatility1, indent=2)
print(volatility_indented)
```

```python
volatility1.sort(key=lambda x: x['Volatility'])
volatility_low_30 = [] 
for share in volatility1[:30]:
    volatility_low_30.append(share['Share'])

nifty100 = pd.read_csv('ind_Nifty100LowVolatility30list.csv')
top30_official = nifty100['Symbol'].tolist()
    
# for share in volatility_low_30:
#   print(share)
print("Our calculated top 30 low volatility shares")
print(sorted(volatility_low_30))
print("Official top 30 low volatility shares")
print(top30_official)
```

```python
correct_prediction = set(volatility_low_30).intersection(set(top30_official))
print("total match", len(correct_prediction))

# while testing it gave 30 on 30 but started it gave 21 on 30 afterwards without any cahnges in the code
```
